# BFS

1. 类型 （使用条件）

二叉树， 图（拓扑排序， 层级遍历， 由点及面）简单图最短路径

2. 能用BFS解决的问题不要用DFS， 除非面试官要求， DFS用recursive实现的DFS可能stackoverflow 而nonrecursive 比较复杂

3. BFS要点

3.1 层级遍历比不分层遍历多一个循环

3.2 BFS时间复杂度 O(e) e是边数

3.3 步骤：

a. 起点 -- 队列

b. while(队列不空){ 

        当前层----> 下一层
        
        for (当前的每一个节点，将下一层放入， 删除当前节点）

}
4. 二叉树序列化

为了能够在重构二叉树时结点能够插入到正确的位置，在使用先序遍历保存二叉树到文件中的时候需要把NULL结点也保存起来（可以使用特殊符号如“#”来标识NULL结点）

5. 二叉树的BFS： 层序遍历

6. 图上的BFS： 由点及面

a. 图中存在环， 需要用hashmap/hashset 判重， 因为环可能使同一个节点重复进入队列

b. 多不需要分层，每次取队列中的一个点即可

题型： graph valid tree: I. N-1 条边  II. 几个点连通
      
      clone graph
      
      search graph nodes
      
      follow up: 如果找所有最近的value == target 的点， 需要分层遍历
      
      
c. 托盘排序

题型： topology sorting，  course schedule,  sequence reconstruction: 判断是否只存在一个拓扑排序的序列， 只需保证队列中一直最只有一个元素

7. 矩阵的BFS

a. 由点及面， 简单图的最短路径

b. 复杂度 r行c列， 边约2rc条

c. 坐标变换数组  is_iland 

d. 题： number of iland, zombie in matrix, knight shortest path



# DFS

1. BFS VS DFS on binary tree

BFS 耗费空间由树宽度的最宽处决定， DFS 耗费的空间是由路径的长度决定路径的长度

以时间换空间的方式： 用DFS模拟BFS时， 相当于有层数限制的搜索，搜到第i层就返回

2. 什么时候用DFS：

找所有方案的题，一定是DFS， 90%的DFS题时排列组合， 排列组合的问题，深度不深，但是方案很多

3. 面试官不特定要求，DFS用递归实现

4. 三种类型题目： 排列， 组合， 图

5. DFS 通用的时间复杂度计算公式： 最好 O(答案个数*每个答案构造时间)； 最坏 O(n!*构造每个答案时间)

6. 组合

题目： subset, combination sum, palindrone partitioning

通常四个参数： vector result; vector subset; int start idx; vector nums

7. 排列： 不再需要start idx, 用hashset判断一个数在不在里面， 将参数 start idx 换为 set(int)

题目： permutation, N-queen

8. 图中搜素：

题目： word ladder

9. 非递归必背： binary tree preoder/inorder/postorder, binary search tree iteration, subset(二进制逐个枚举)， permutaions

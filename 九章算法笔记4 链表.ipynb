```python
class ListNode:
    
    def __init__(self, data=0, next_node=None):
        self.data = data
        self.next = next_node
        
def search_list(L, key):
    while L and L.data != key:
        L = L.next
    return L

def insert_after(node, new_node):
    new_node.next = node.next # order is important since you can not lose the node.next!!!
    node.next = new_node
    
def delete_after(node):
    if node.next:
        node.next = node.next.next
    
def delete_from_list(L, node):
    if not node:
        return
    if node.next: # node is not tail
        node.data = node.next.data
        node.next = node.next.next
    else:
        dummy_head = ListNode(0, L)
        pre, cur = dummy_head, L
        while cur.next:
            pre, cur = pre.next, cur.next
        pre.next = None
    
def remove_kth_last(L, k):# assume L has at least k nodes
    dummy_head = ListNode(0, L)
    first = dummy_head.next
    for _ in range(k):
        first = first.next
    second = dummy_head     # 0 --- k+1
    while first:
        first, second = first.next, second.next
    second.next = second.next.next
    return dummy_head.next

# you can also remove kth node prior to a given node by using the same idea --- advancing + while check
           
```

```python
def has_cycle(L):
    fast, slow = L, L
    while fast and fast.next and fast.next.next:
        slow, fast = slow.next, fast.next.next
        if slow is fast:
            slow = L
            while slow != fast:
                slow, fast = slow.next, fast.next
            return slow
    return None # no cycle
def cycle_length(L):
    entrance = has_cycle(L)
    if not entrance:
        return 0
    count = 1
    temp = entrance.next
    while temp != entrance:
        temp = temp.next
        count += 1
    return count

a = ListNode(1)
b = ListNode(2)
c = ListNode(3)
d = ListNode(4)
e = ListNode(5)
f = ListNode(6)
a.next = b
b.next = c
c.next = d
d.next = e
e.next = f
f.next = c

print(has_cycle(a).data)
print(cycle_length(a))
```

5.4 merge sort更简单

step 1：findMiddle（快慢指针）；//对数据流问题，不可重新扫，数据实时到来；

step 2：merge two list；

step 3：总框架 { 找中点 -> 左排序（递归） ->右排序（递归） -> 合并 }

5.5 代码
！！！注：
findMiddle函数中，如果fast和slow都从head开始，则 while(fast->next && fast->next->next)
如果slow从head开始，fast从head的下一个开始，则 while(fast && fast->next)

举例：1 -> -1 ->null。mid的位置必须在1，而不是-1，因为right的排序是从mid->next开始的。


```python
def findMiddle(L):
    if not L:
        return None
    slow, fast = L, L
    while fast and fast.next and fast.next.next: # while fast.next and fast.next.next is okay
        slow, fast = slow.next, fast.next.next
    return slow

def merge(L1, L2):
    if not L1:
        return L2
    if not L2:
        return L1
    dummy_head = ListNode(0)
    head, head1, head2 = dummy_head, L1, L2
    while head1 and head2:
        if head1.data < head2.data:
            head.next = head1
            head1 = head1.next
        else:
            head.next = head2
            head2 = head2.next
        head = head.next  #!!!
    if head1:
        head.next = head1
    if head2:
        head.next = head2
    return dummy_head.next

def sortList(L):
    if not L or not L.next:
        return L
    middle = findMiddle(L)
    right = sortList(middle.next)
    middle.next = None
    left = sortList(L)
    return merge(left, right)

# corner case 的处理都很关键
a = ListNode(1)
b = ListNode(3)
c = ListNode(2)
d = ListNode(4)
e = ListNode(6)
f = ListNode(5)
a.next = b
b.next = c
c.next = d
d.next = e
e.next = f
f.next = None  

sortList(a)
head = a
while head:
    print(head.data)
    head = head.next
```
